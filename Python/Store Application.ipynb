```python
import random
from prettytable import PrettyTable

# Step 1: Set up predefined user credentials
user = "Test".lower()
password = "1111"

# Step 2: Login and verification loop
while True:
    x = input('Enter your username: ').lower().strip()
    if x == user:
        y = input('Enter your password: ')
        if y == password:
            s = random.randrange(1000, 10000)  # Generate random verification code
            print("Your verification code is:", s)
            while True:
                try:
                    i = int(input('Enter your verification code: '))
                    if i == s:
                        print("Verification successful.\nWelcome to our application!\n")
                        break
                    else:
                        print("Incorrect verification code. Try again.")
                except ValueError:
                    print("Please enter a valid numeric verification code.")
            break
        else:
            print("Incorrect password. Try again.")
    else:
        print("Incorrect username. Try again.")

# Step 3: Define available products
products = [
    {"name": "Water", "price": 80.0, "quantity": 1200},
    {"name": "Soda", "price": 100.0, "quantity": 800},
    {"name": "Juice", "price": 120.0, "quantity": 600},
    {"name": "Coffee", "price": 100.0, "quantity": 400},
    {"name": "Ice Cream", "price": 150.0, "quantity": 200},
]

# Step 4: Display products in a table format
table1 = PrettyTable()
table1.field_names = ["Product Name", "Price", "Quantity"]
for product in products:
    table1.add_row([product["name"], product["price"], product["quantity"]])
print("Available Products:")
print(table1)

# Step 5: Initialize totals
total_amount = 0
total_discount = 0

# Create a lowercase product name list using map (for validation)
product_names_lower = list(map(lambda x: x["name"].lower(), products))

# Step 6: Start ordering loop
while True:
    product_name = input("Enter the product name you want to buy (or type 'exit' to finish): ").strip().lower()
    if product_name == "exit":
        break

    # Find the product
    product = next((p for p in products if p["name"].lower() == product_name), None)
    if not product:
        print("Product not found. Please try again.")
        continue

# Get product from original list
    product = next(filter(lambda x: x["name"].lower() == product_name, products), None)

    # Validate quantity input
    while True:
        quantity_input = input(f"Enter the quantity of {product['name']}: ")
        if quantity_input.isdigit():
            quantity = int(quantity_input)
            break
        else:
            print("Please enter a valid number for quantity.")

    # Check stock availability
    if quantity > product["quantity"]:
        print(f"Insufficient stock. Only {product['quantity']} available.")
        continue

    # Step 7: Calculate prices and discount
    product["quantity"] -= quantity
    total_price = product["price"] * quantity
    discount = 0.05 * total_price if quantity >= 10 else 0
    total_amount += total_price
    total_discount += discount

    final_price = total_price - discount
    print(f"You ordered {quantity} {product['name']}(s).")
    print(f"Total price: {total_price} | Discount: {discount} | Final price: {final_price}\n")

# Step 8: Show final receipt
print("\n--- Receipt ---")
print(f"Total price before discount: {total_amount}")
print(f"Total discount received: {total_discount}")
print(f"Total price after discount: {total_amount - total_discount}")
print("Thank you for shopping with us!")
```

**Creating a Full class from the previous Application:**

```python
import random
from prettytable import PrettyTable

class ShoppingApp:
    def __init__(self):
        # Initialize user credentials and product list
        self.user = "test"
        self.password = "1111"
        self.products = [
            {"name": "Water", "price": 80.0, "quantity": 1200},
            {"name": "Soda", "price": 100.0, "quantity": 800},
            {"name": "Juice", "price": 120.0, "quantity": 600},
            {"name": "Coffee", "price": 100.0, "quantity": 400},
            {"name": "Ice Cream", "price": 150.0, "quantity": 200},
        ]
        self.total_amount = 0
        self.total_discount = 0

    def login(self):
        # User login and verification code check
        while True:
            username = input("Enter your username: ").lower()
            if username == self.user:
                pwd = input("Enter your password: ")
                if pwd == self.password:
                    code = random.randrange(100, 999)
                    print(f"Your verification code is: {code}")
                    while True:
                        try:
                            user_code = int(input("Enter your verification code: "))
                            if user_code == code:
                                print("Verification successful! Welcome to the app.\n")
                                return True
                            else:
                                print("Incorrect verification code. Try again.")
                        except ValueError:
                            print("Please enter a valid number.")
                else:
                    print("Incorrect password. Try again.")
            else:
                print("Incorrect username. Try again.")

    def show_products(self):
        # Display product catalog
        table = PrettyTable()
        table.field_names = ["Product Name", "Price", "Quantity"]
        for product in self.products:
            table.add_row([product["name"], product["price"], product["quantity"]])
        print("Available Products:")
        print(table)

    def find_product(self, product_name):
        # Search for a product by name (case-insensitive)
        return next((p for p in self.products if p["name"].lower() == product_name.lower()), None)

    def order(self):
        # Handle ordering process
        while True:
            product_name = input("Enter the product name to buy (or type 'exit' to finish): ").strip()
            if product_name.lower() == "exit":
                break

            product = self.find_product(product_name)
            if not product:
                print("Product not found. Please try again.")
                continue

            quantity_input = input(f"Enter quantity of {product['name']}: ")
            if not quantity_input.isdigit():
                print("Please enter a valid number.")
                continue

            quantity = int(quantity_input)
            if quantity > product["quantity"]:
                print(f"Not enough stock. Available: {product['quantity']}")
                continue

            # Update quantities and totals
            product["quantity"] -= quantity
            price = product["price"] * quantity
            discount = 0.05 * price if quantity >= 10 else 0
            self.total_amount += price
            self.total_discount += discount

            print(f"You ordered {quantity} {product['name']}(s).")
            print(f"Total price: {price} | Discount: {discount} | Final price: {price - discount}\n")

    def show_receipt(self):
        # Display final receipt
        print("\n--- Receipt ---")
        print(f"Total before discount: {self.total_amount}")
        print(f"Total discount: {self.total_discount}")
        print(f"Total after discount: {self.total_amount - self.total_discount}")
        print("Thank you for shopping with us!")

    def run(self):
        # Main control function
        if self.login():
            self.show_products()
            self.order()
            self.show_receipt()

# Run the application
app = ShoppingApp()
app.run()
```
